```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import fdringdown as rd
```

```python
wf = rd.waveforms.ringdown(modes=[(2,2,0),(3,3,0)])
```

```python
times = np.linspace(-0.01, 0.05, num=2000)

params_pyring = {
    'A_rd_0': 1.,
    'A_rd_1': 0.5,
    'phi_rd_0': 0.2,
    'phi_rd_1': 0.4,
    'iota': np.pi/6,
    'varphi': 0,
    't_0': 0,
    'M_f': 70,
    'chi_f': 0.7
}

pyring_wf = wf.waveform(times, params_pyring)
```

```python
fig, ax = plt.subplots(dpi=180)

ax.plot(times, np.real(pyring_wf))

for phi0 in np.linspace(-0.5, 0.5, 10)[3:4]:
    
    print(phi0)
    
    for phi1 in np.linspace(-0.5, -0.3, 5)[2:3]:
        
        print(phi1)
    
        params_isi = {
            'A_rd_0': 1.,
            'A_rd_1': 0.5,
            'phi_rd_0': phi0,
            'phi_rd_1': phi1,
            'iota': np.pi/6,
            'varphi': 0,
            't_0': 0,
            'M_f': 70,
            'chi_f': 0.7
        }
    
        isi_wf = wf.waveform_isi(times, params_isi)

        ax.plot(times, np.real(isi_wf), c='C1', alpha=1, linestyle='--')
    
ax.set_xlim(0.005, 0.015)
ax.set_ylim(-0.2, 0.2)
```

```python
fig, ax = plt.subplots(dpi=180)

ax.plot(times, np.imag(pyring_wf))
ax.plot(times, np.imag(isi_wf))
```
